# Audio Features Extraction 

## Overview

In this notebook we will be extracting from our audio datasets (original and the sigmented one) a variety of features such as *tempo*, *spectral characteristics*, and *Mel-frequency*, which will serve as the basis for classification.

```python
import librosa
from glob import glob
import os
import numpy as np
import pandas as pd
import csv
```

## features extraction function

*This function aims to extract the features from the audio files then store them into a csv file*

```python
def features_extraction(input_directory, csv_file):
    header = [
        'filename', 'tempo', 'mean_y_harmonic', 'var_y_harmonic', 'mean_y_perceptr','var_y_perceptr',
        'mean_rms', 'var_rms', 'mean_zcr', 'var_zcr', 'mean_spectral_cent',
        'var_spectral_cent', 'mean_spectral_rolloff', 'var_spectral_rolloff',
        'mean_spectral_bandwidth', 'var_spectral_bandwidth', 'mean_chroma_stft',
        'var_chroma_stft'
    ]

    for i in range(1, 21):
        header.append(f'mfcc_mean{i}')

    for i in range(1, 21):
        header.append(f'mfcc_var{i}')

    header.append('label')

    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)

    audio_files = glob(os.path.join(input_directory, '**/*.wav'), recursive=True)

    for file_path in sorted(audio_files):
        try:
            y, sr = librosa.load(file_path)
            
            y_harmonic, y_perceptr = librosa.effects.hpss(y)
            
            filename = os.path.basename(file_path)

            tempo = features_computation.tempo(y_harmonic, sr)

            zcr = features_computation.zero_crossing_rate(y=y)

            rms = features_computation.rms(y=y)

            spectral_cent = features_computation.spectral_centroid(y=y, sr=sr)

            spectral_rolloff = features_computation.spectral_rolloff(y=y, sr=sr)

            spectral_bandwidth = features_computation.spectral_bandwidth(y=y, sr=sr)

            chroma_stft = features_computation.chroma(y=y, sr=sr)

            mfcc = features_computation.mfcc(y=y, sr=sr)
            
            filename_without_extension = os.path.splitext(filename)[0]
            label = filename_without_extension.split(".")[0]

            features_list = [
                filename, tempo, np.mean(y_harmonic), np.var(y_harmonic), np.mean(y_perceptr),
                np.var(y_perceptr), np.mean(rms), np.var(rms), np.mean(zcr), np.var(zcr),
                np.mean(spectral_cent), np.var(spectral_cent), np.mean(spectral_rolloff),
                np.var(spectral_rolloff), np.mean(spectral_bandwidth),
                np.var(spectral_bandwidth), np.mean(chroma_stft), np.var(chroma_stft)
            ]

            for e_mean in mfcc:
                features_list.append(np.mean(e_mean))

            for e_var in mfcc:
                features_list.append(np.var(e_var))

            features_list.append(label)

            with open(csv_file, 'a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(features_list)

            print(f"File: {file_path} processed")

        except (librosa.util.exceptions.ParameterError, FileNotFoundError) as e:
            print(f"Error processing {file_path}: {e}")
```

## Extract the 30 secondes audio features

```python
input_directory = 'dataset/Audio_30s'
output_csv_file = 'dataset/data_30sc.csv'
features_extraction(input_directory, output_csv_file)
```
